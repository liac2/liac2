```python
import yfinance as yf
tickers = ["AAPL"]
a = yf.download(tickers, start="2010-01-01", end="2021-01-01")
a
```

```python
a = a["Close"]
a
```

```python
a["change"] = a.a.pct_change(periods=1).mul(100)
```

```python
a.rename(columns={"a": "data"}, inplace=True)
a
```
